```python
data = core.generate_2D_sample(300, 10, 200, 800)
print(data[0].shape)
print(data[1].shape)
```

```python
x = data[0]
y = data[1]

print(y.shape)

# plotting a single vector
d0 = y[:,0]

# Plot the generated 1D vector
plt.plot(x, d0)
plt.title('Random 1D Vector with 3 Gaussians and White Noise')
plt.xlabel('X-axis')
plt.ylabel('Values')
plt.grid()
plt.show()
```

```python
U, S, Vh = np.linalg.svd(y, full_matrices=True)

n=10
offset = 0.275
for i in range(n):
    plt.plot(U[:,i]+offset*i)

plt.grid()
plt.title('Eigenvectors (from bottom to top)')
plt.xlabel('X-axis')
```

```python
# see eigenvalues

n = 23
view = 10
subset =  S[:n]
minY  = np.min(subset)
maxY = np.max(subset)

plt.plot(S, 'o-')
plt.yscale('log')
plt.ylim([minY, maxY])
plt.xlim([0, view])
plt.grid()

 
plt.title('Eigenvalues')
plt.xlabel('index')
```
